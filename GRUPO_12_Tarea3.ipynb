```python
#1. Probabilidad de obtener exactamente 7 caras en 10 lanzamientos
#Usamos la funci칩n binom.pmf porque tenemos una situaci칩n que tiene dos resultados posibles adem치s de que se repetir치 cierto n칰mero de veces
from scipy.stats import binom
n = 10
k = 7
p = 0.5
probabilidad = binom.pmf(k, n, p)
probabilidad
```

```python
#2.Calcula la probabilidad acumulada de obtener 7 o menos caras en 10 lanzamientos
#La funci칩n binom.cdf nos permitir치 calcular esa probabilidad acumulada
from scipy.stats import binom
n = 10
k = 7
p = 0.5
probabilidad_acum = binom.cdf(k, n, p)
probabilidad_acum
```

```python
#3. Grafica la distribuci칩n binomial completa para los 10 lanzamientos, mostrando la probabilidad de obtener desde 0 hasta 10 caras.
import matplotlib.pyplot as plt
import numpy as np

# Definimos los par치metros
n = 10  # n칰mero de lanzamientos
p = 0.5 # probabilidad de 칠xito

# Valores posibles de caras (de 0 a 10), np.arrange(a,b) nos dar치 un vector que comienza en a y termina en b-1
val_posibles = np.arange(0, n+1)

# Calculamos las probabilidades usando la funci칩n binom.pmf
probabilidades = binom.pmf(val_posibles, n, p)

# Graficamos la distribuci칩n binomial
plt.bar(val_posibles, probabilidades, color='#00008B') #generamos un gr치fico de barras con los datos de la distribuci칩n normal
plt.title('Distribuci칩n Binomial de 10 Lanzamientos de una Moneda') #t칤tulo del gr치fico
plt.xlabel('N칰mero de Caras') #nombre del eje x
plt.ylabel('Probabilidad') #nombre del eje y
plt.xticks(val_posibles) #valores que aparecer치n en el eje x
plt.grid(axis='y', linestyle='--', alpha=0.7) #colocamos un grid en el gr치fico y le damos formato
plt.show()
```

```python
#5. Utilizando 洧눖洧눇洧눍洧눔洧눜. 洧눖洧눗洧눅洧눗洧눖, realiza un an치lisis de una distribuci칩n normal con una media de 0 y desviaci칩n est치ndar de 1. Genera un conjunto de datos aleatorios con 1000
#muestras a partir de esta distribuci칩n.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

#Asignamos los valores de los par치metros de la distribuci칩n normal
media = 0
dev = 1 #desviaci칩n est치ndar
nro_muestras = 1000

#generamos los datos aleatoriamente como nos pide el enunciado
datos = np.random.normal(media, dev, nro_muestras)

print(datos[:100])
```

```python
#6. Calcula y muestra la media, la desviaci칩n est치ndar y la mediana de los datos generados
media_muestral = np.mean(datos)
dev_estandar_muestral = np.std(datos)
mediana_muestral = np.median(datos)

media_muestral, dev_estandar_muestral, mediana_muestral
```

```python
#7. Grafica el histograma de los datos y superp칩n la curva de la distribuci칩n normal te칩rica.

# Graficamos el histograma de los datos generados
plt.hist(datos, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

# Establecemos valores para x en la distribuci칩n normal te칩rica
x = np.linspace(min(datos), max(datos), 100)

# Calcular la densidad de la distribuci칩n normal te칩rica
pdf = stats.norm.pdf(x, media, dev) #recordemos que media = 0 y dev = 1

# Graficamos la distribuci칩n normal te칩rica
plt.plot(x, pdf, 'k', linewidth=2)
plt.title('Datos generados y curva de la distribuci칩n normal te칩rica')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.show()
```

```python
#8. Calcula la probabilidad de que un valor caiga dentro del rango de -1 a 1 (una desviaci칩n est치ndar de la media)
#Para calcular la probabilidad necesitamos hallar la funci칩n de distribuci칩n acumulativa (CDF) para obtener la diferencia entre las funciones que se asignen a 1 y -1
import scipy.stats as stats

prob1= stats.norm.cdf(1, loc=media, scale=dev)
prob2= stats.norm.cdf(-1, loc=media, scale=dev)

probabilidad8 = prob1-prob2
probabilidad8
```
